## Combining both the Data sets

```python
app_data1 = app_data[[col for col in app_data.columns if col not in ["TARGET"]]]
prev_data1= prev_data[[col for col in prev_data.columns if col not in ["TARGET"]]]
app_data1["Type"] = "application_data"
prev_data1["Type"] = "previous_application"
new_data = pd.concat([app_data1, prev_data1], axis=0)
```

### Distribution of contract type ,gender ,own car ,own house with respect to Loan Repayment status

```python
defaulter = app_data[app_data["TARGET"]==1][['NAME_CONTRACT_TYPE', 'CODE_GENDER','FLAG_OWN_CAR', 'FLAG_OWN_REALTY']]
non_defaulter = app_data[app_data["TARGET"]==0][['NAME_CONTRACT_TYPE', 'CODE_GENDER','FLAG_OWN_CAR', 'FLAG_OWN_REALTY']]

cols = ['NAME_CONTRACT_TYPE', 'CODE_GENDER','FLAG_OWN_CAR', 'FLAG_OWN_REALTY']

plt.figure(figsize=[20,5])

import itertools

for i,j in itertools.zip_longest(cols,range(len(cols))):
    plt.subplot(1,4,j+1)
    defaulter[i].value_counts().plot.pie(autopct = "%1.0f%%",startangle = 90,)
    circ=plt.Circle((0,0),.7,color="white")
    plt.gca().add_artist(circ)
    plt.ylabel("")
    plt.title(i+"-Defaulter")


plt.figure(figsize=(20,5))
for i,j in itertools.zip_longest(cols,range(len(cols))):
    plt.subplot(1,4,j+1)
    non_defaulter[i].value_counts().plot.pie(autopct = "%1.0f%%",startangle = 90)
    circ=plt.Circle((0,0),.7,color="white")
    plt.gca().add_artist(circ)
    plt.ylabel("")
    plt.title(i+"-Repayer")
```
